## Матриця яскравості (RGB)

```python
rows, cols, _ = img.shape

brightness_matrix = [[0 for x in range(cols)] for y in range(rows)] 

for i in range(rows):
    for j in range(cols):
        brightness_matrix[i][j] = sum(img[i, j]) / 3
```

### Функція для вибірки вікна

```python
import math
```

```python
def get_index(array, i, j):
    try:
        return array[i][j]
    except:
        return 0
```

```python
# i - starting row
# j - starting column
# size can only be odd and >= 3
def window3x3(matrix, i, j):
    result = np.zeros((3, 3))
    
    # root
    result[1][1] = get_index(matrix, i, j)
    
    # north
    result[0][1] = get_index(matrix, i - 1, j)
    
    # north east
    result[0][2] = get_index(matrix, i - 1, j + 1)
        
    # north west
    result[0][0] = get_index(matrix, i - 1, j - 1)
    
    # east
    result[1][2] = get_index(matrix, i, j + 1)
        
    # south
    result[2][1] = get_index(matrix, i + 1, j)
        
    # south east
    result[2][2] = get_index(matrix, i + 1, j + 1)
        
    # south west
    result[2][0] = get_index(matrix, i + 1, j - 1)
        
    # west
    result[1][0] = get_index(matrix, i, j - 1)
    
    return result
            
```

## Фільтрація методом осереднення ковзним вікном з використанням вагових коефіцієнтів для значень сусідніх пікселів.

```python
img_noise_copy = np.copy(img_noise)
noise_reduction_coefficient = 0.25
```

```python
for i in range(img_noise_copy.shape[0]):
    for j in range(img_noise_copy.shape[1]):
        img_noise_copy[i][j] = noise_reduction_coefficient * np.array(window3x3(img_noise, i, j)).mean()
```

```python
plt.imshow(cv2.cvtColor(img_noise_copy, cv2.COLOR_BGR2RGB))
plt.show()
```
